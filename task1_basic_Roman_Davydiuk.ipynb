Завдання 2: Розв’язати квадратне рівняння $ax^2+bx + c = 0$.  Напишіть функцію,яка приймає параметри квадратного рівняння, а повертає одне рішення або декілька.

Наприклад:  


```
solve_quad(a=1, b=-2, c=1) -> 1  
solve_quad(a=1, b=0, c=-1) -> (-1, 1)  
solve_quad(a=0, b=2, c=-1) -> 0.5
```







````python
import cmath

def solve_quad(a, b, c):
    
    if a == 0:
        raise ValueError("Коефіцієнт 'a' не може бути нулем.")
    
    
    discriminant = (b**2) - (4*a*c)

    
    if discriminant > 0:
        
        x1 = (-b + cmath.sqrt(discriminant)) / (2*a)
        x2 = (-b - cmath.sqrt(discriminant)) / (2*a)
        return x1, x2
    elif discriminant == 0:
        
        x = -b / (2*a)
        return x
    else:
        
        x1 = (-b + cmath.sqrt(discriminant)) / (2*a)
        x2 = (-b - cmath.sqrt(discriminant)) / (2*a)
        return x1, x2

print(solve_quad(a=1, b=-2, c=1))  # 1
print("1")
print(solve_quad(a=1, b=0, c=-1))  # (-1, 1)
print("2")
````

Завдання 3: Реалізація функції merge. На вхід функція приймає 2 відсортованих масиви - `a` і `b`. Функція повинна повернути один відсортований масив, який містить всі елементи масивів `a` і `b`.

Наприклад: `merge([1, 3, 3, 5], [3, 5, 5, 7]) -> [1, 3, 3, 3, 5, 5, 5, 7]`

````python
from typing import List

def merge(a: List[int], b: List[int]) -> List[int]:
    
    result = a + b

    
    result.sort()

    
    i = 0
    while i < len(result) - 1:
        if result[i] == result[i + 1]:
            del result[i]
        else:
            i += 1

    return result


print(merge([1, 3, 3, 5], [3, 5, 5, 7]))  
````

Завдання 4: Реалізація алгоритму [MergeSort](https://ru.wikipedia.org/wiki/%D0%A1%D0%BE%D1%80%D1%82%D0%B8%D1%80%D0%BE%D0%B2%D0%BA%D0%B0_%D1%81%D0%BB%D0%B8%D1%8F%D0%BD%D0%B8%D0%B5%D0%BC)..

Наприклад: `MergeSort([3, 1, 3, 5]) -> [1, 3, 3, 5]`

````python
def merge_sort(arr: List[int]) -> List[int]:
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]

    left = merge_sort(left)
    right = merge_sort(right)

    return merge(left, right)


print(merge_sort([3, 1, 3, 5]))  
````

Завдання 6: Написати функцію яка на вхід приймає 3 масиви – `X, Y, Z`. функція повертатиме `True` у випадку, коли вони рівні та `False` – якщо ні.

Рекомендація до завдання: Не використовуйте оператор `==`

````python
import numpy as np

def arrays_equal(arr1: np.ndarray, arr2: np.ndarray, arr3: np.ndarray) -> bool:
    
    arr1 = np.array(arr1)
    arr2 = np.array(arr2)
    arr3 = np.array(arr3)

    
    return arr1.shape == arr2.shape == arr3.shape and np.all(arr1 == arr2) and np.all(arr2 == arr3)

arr1 = [1, 2, 3]
arr2 = [1, 2, 3]
arr3 = [1, 2, 4]
arr4 = [1, 2, 3, 4]

print(arrays_equal(arr1, arr2, arr3))  
print(arrays_equal(arr2, arr3, arr4))  
print(arrays_equal(arr1, arr2, arr2))  
````

Завдання 7: написати функцію, яка приймає на вході вектор рядок, а повертає вектор стовбець.

````python
import numpy as np

def rot_vec(x: np.ndarray) -> np.ndarray:
   
    if x.ndim != 1:
        raise ValueError("Input should be a 1D array")

  
    return x.reshape(-1, 1)


arr = np.array([1, 2, 3, 4])
print(rot_vec(arr))
````

Завдання 8: Написати функцію яка приймає на вхід двовимірний масив, а повертає масив з діагональними елементами цієї матриці.

Наприклад: `[[1, 2], [3, 4]] -> [1 4]`

````python
import numpy as np

def get_diag(X: np.ndarray) -> np.ndarray:
    
    if X.ndim != 2:
        raise ValueError("Input should be a 2D array")

    
    return np.diagonal(X)

arr = np.array([[1, 2], [3, 4]])
print(get_diag(arr))
````

Завдання 9: Написати функцію яка приймає на вхід двомірний масив, а повертає масив елементів які знаходяться на другорядній діагоналі. Приклад: `[[1, 2], [3, 4]] -> [2 3]`

````python
import numpy as np

def sec_diag(X: np.ndarray) -> np.ndarray:
    
    if X.ndim != 2:
        raise ValueError("Input should be a 2D array")


    return X[::-1, :].diagonal()


arr = np.array([[1, 2], [3, 4]])
print(sec_diag(arr))
````

Завдання 12: Для наведених та реалізованих графіків Вам нобхідно зробити розширення (додати підписи, легенду тощо):

$f_{1}(x) =  \frac{sin(x)}{x}$

$f_{2}(x) =  \frac{cos(x)}{x^2 + 1}$

$f_{3}(x) =  cos^2(x) - \frac{sin(x)}{2}$  

$f_{4}(x) =  cos(x)$  

$f_{5}(x) =  \frac{sin(x)}{x^2 + 1}$

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000)

# f1(x) = sin(x)/x
f1 = np.sin(x)/x

# f2(x) = cos(x)/(x^2 + 1)
f2 = np.cos(x)/(x**2 + 1)

# f3(x) = cos^2(x) - sin(x)/2
f3 = np.cos(x)**2 - np.sin(x)/2

# f4(x) = cos(x)
f4 = np.cos(x)

# f5(x) = sin(x)/(x^2 + 1)
f5 = np.sin(x)/(x**2 + 1)

fig, ax = plt.subplots()

ax.plot(x, f1, label='f1(x) = sin(x)/x')
ax.plot(x, f2, label='f2(x) = cos(x)/(x^2 + 1)')
ax.plot(x, f3, label='f3(x) = cos^2(x) - sin(x)/2')
ax.plot(x, f4, label='f4(x) = cos(x)')
ax.plot(x, f5, label='f5(x) = sin(x)/(x^2 + 1)')

ax.set_title('Plot of f1(x), f2(x), f3(x), f4(x), and f5(x)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()

plt.show()
````
